```python
# https://stackoverflow.com/a/1176023
def to_snake(s: str):
    pattern = re.compile(r"(?<=[a-z])(?=[A-Z])|(?<=[A-Z])(?=[A-Z][a-z])")
    return pattern.sub('_', s).lower()

REPLACE_TYPES = {
    "Int": "int",
    "Int64": "int",
    "Bool": "bool",
    "ByteString": "bytes",
    "Text": "str",
    "String": "str",
}

REPLACE_NAMES = {
    "ChatDeleteMode": "mode",
    "AppSettings": "settings",
    "ChatSettings": "settings",
    "ArchiveConfig": "config",
    "DBEncryptionConfig": "config",
    "AppFilePathsConfig": "config",
    "NetworkConfig": "config",
    "SimpleNetCfg": "config",
    "IncognitoEnabled": "incognito",
    "CreateShortLink": "short",
}
def parse_type(t: str):
    if t[0] == "(":
        assert t[-1] == ")", t
        return parse_type(t[1:-1])
    if t.startswith("Maybe "):
        a,b = parse_type(t.split(" ",1)[1])
        return f"Optional[{a}]", b
    if t.startswith("NonEmpty "):
        a,b = parse_type(t.split(" ",1)[1])
        return f"list[{a}] # NonEmpty", b
    if t.startswith("["):
        assert t[-1] == "]", t
        a,b = parse_type(t[1:-1])
        return f"list[{a}]", b
    if t in REPLACE_TYPES:
        t = REPLACE_TYPES[t]
    return t, t

name_types = []
for cp in commands_parse:
    print(f"class {cp[0]}(BaseChatCommand):")
    for attr in cp[1]:
        if " :: " in attr:
            name, type_ = attr.split(" :: ")
            name = to_snake(name)
        else:
            type_ = attr
            type_, underlying = parse_type(type_)
            if underlying in REPLACE_NAMES:
                underlying = REPLACE_NAMES[underlying]
            name = to_snake(underlying)
        #if (name, type_) not in name_types:
        #    name_types.append((name, type_))
        print(f"    {name}: {type_}")
    print("    def format(self) -> str:")
    for p in cp[2]:
        p = p.replace("sendMsgQuote", "SendMessageQuote")
        p = p.replace(cp[0], "")
        p = p.replace("<*>", "+").replace("*>", "+").replace("<*", "+")
        p = p.replace("<$>", "").replace("$>", "").replace("<$", "")
        print(f"        return {p}")
    print()
#unique_types = set(t for (n,t) in name_types)
#for t in sorted(unique_types): print(t)
```
